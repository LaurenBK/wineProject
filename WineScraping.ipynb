```python
dat = []

def findInfo(soup):
    d = [] 
    attr = ['WineMaker','Variety','Origin','Alcohol','pH','ResidualSugar','TA']
    name = 'lit_WineName'
    notes = 'TastingNotes'
    ageing = 'Ageing'
    blend = 'BlendPercent'
    d.append(soup.find('span', id=name).text.strip())
    for a in attr:
        try:
            name_box = soup.find('div', id='pnl_'+a)  
            name = name_box.text.strip()
            d.append(name.split('\r\n\t\t\t\t\t                                    ')[1])
        except:
            d.append(np.nan)
    try:
        d.append(soup.find('div', id='Pnl_'+notes).text.strip().split('Tasting Notes\n')[1])
    except:
        d.append(np.nan)
    try:
        d.append(soup.find('div', id='Pnl_'+ageing).text.strip().split('Ageing Potential\n')[1])
    except:
        d.append(np.nan)
    try:
        d.append(soup.find('div', id='Pnl_'+blend).text.strip().split('Blend Information\n')[1])
    except:
        d.append(np.nan)
    return d

dat.append(findInfo(soup))
data = pd.DataFrame(dat, columns = ['Name','WineMaker','Variety','Origin','Alcohol (%vol)','pH','ResidualSugar (g/l)','TA (g/l)','TastingNotes','Ageing','BlendPercent'])
```

```python
def collectData(urlNum):
    soups = []; names = []
    for n in urlNum:
        try:
            url = 'http://www.wine.co.za/wine/wine.aspx?WINEID='+str(n)
            page = u2.urlopen(url).read()
            soup = BeautifulSoup(page,'html.parser')
            try: 
                name = soup.find('span', id='lit_WineName').text.strip()
                names.append(name)
                soups.append(soup)
            except:
                soups.append('')
        except:
            print 'Number',n,'not working'
            soups.append('')
        sleeptime = np.random.random()+0.5 
        print sleeptime
        time.sleep(sleeptime)
    return soups,names
    
urlNum = range(5840,5860)
def collectionOfData(urlRange): 
    soups,names = collectData(urlRange)
    print 'Names in this batch:',names
    dat = []
    for s in soups:
        if s == '':
            pass
        else:
            dat.append(findInfo(s))
    print len(dat)
    data = pd.DataFrame(dat, columns = ['Name','WineMaker','Variety','Origin','Alcohol (%vol)','pH','ResidualSugar (g/l)','TA (g/l)','TastingNotes','Ageing','BlendPercent'])
#     return data
    data.to_csv('WineData/wineData_'+str(urlRange[0])+'_'+str(urlRange[-1])+'.csv', columns = data.columns, encoding='utf-8')
```
